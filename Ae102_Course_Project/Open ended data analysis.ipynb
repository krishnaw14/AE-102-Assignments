```python
Crime_Heads = ["RAPE", "KIDNAPPING AND ABDUCTION", "DOWRY DEATHS", "ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY",
     "INSULT TO THE MODESTY OF WOMEN", "CRUELTY BY HUSBAND OR HIS RELATIVES (IPC SECTION 498A)","IMPORTATION OF GIRLS FROM FOREIGN COUNTRY", "IMMORAL TRAFFIC (P) ACT", "DOWRY PROHIBITION ACT", 
     "INDECENT REPRESENTATION OF WOMEN (P) ACT", "COMMISSION OF SATI (P) ACT", "TOTAL CRIMES AGAINST WOMEN"]
States = list(data1["STATE/UT"][:38])
Year = list(data1)[2:]
```

```python
def f(State,Crime_Head,Year):
    a = list(data1["CRIME HEAD"]).index(Crime_Head)
    b = 0
    for i in range(38):
        if (data1["STATE/UT"][a+i] == State):
            b = a+i
            break
    Year1 = np.arange(2001,2013)
    Values = list(data1.iloc[b])[2:]
    
interact(f,State=Dropdown(options = States, description = "STATE : " ),Crime_Head =Dropdown(options = Crime_Heads , description = "CRIME : " ),Year = fixed(Year))
```

### Predictions for different type of crimes in India (total )

```python
Year1 = np.arange(2001,2013)
predict= [[],[]]
data_2012 = []
for i in range(37,419,38):
    Values = list(data1.iloc[i])[2:]
    linear = np.polyfit(Year1,Values,1)
    data_2012.append(data1["2012"][i])
    value_2013 = linear[0]*2013 + linear[1]
    value_2014 = linear[0]*2014 + linear[1]
    predict[0].append(int(value_2013))
    predict[1].append(int(value_2014))
```

```python
predictions = pd.DataFrame({'Type of Crime' : Crime_Heads[:-1], '2012' : data_2012, '2013' : predict[0], '2014' : predict[1]})
predictions.index += 1
cols = predictions.columns.tolist()
cols = cols[-1:] + cols[0:-1]
predictions[cols]
```
